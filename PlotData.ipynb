```python
%reset -f
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'data_log.txt'  
df = pd.read_csv(file_path)
df['Time'] -= df['Time'].iloc[0]
print("Available options:")
for col in list(df.columns)[1:]:
    print(col)
```

```python
col1 = input("Enter the first column to plot against Time: ")
col2 = input("Enter the second column to plot against Time: ")
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# First subplot
axes[0].plot(df['Time'], df[col1], label=col1, color='b')
axes[0].set_ylabel(col1)
axes[0].legend()
axes[0].grid()

# Second subplot
axes[1].plot(df['Time'], df[col2], label=col2, color='r')
axes[1].set_xlabel('Time (s)')
axes[1].set_ylabel(col2)
axes[1].legend()
axes[1].grid()

plt.suptitle("Time vs Selected Data Columns")
plt.show()
```

```python
print(f"The max value of plot 1 is {max(df[col1])}")
print(f"The max value of plot 2 is {max(df[col2])}")
```
